## Using string split method to get (Date,month and year)

```python
df["Date"]=df["Date_of_Journey"].str.split('/').str[0]
df["Month"]=df["Date_of_Journey"].str.split('/').str[1]
df["Year"]=df["Date_of_Journey"].str.split('/').str[2]
```

```python
df.head(3)
```

```python
# Dropping the Date_of_Journey column

df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
# converting the columns into int type
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
# Splitting Arrival_Time using string split method

df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
# Dropping the Arrival_Time column
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
# converting the columns into int type
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
```

```python
df.head(3)
```

```python
# Splitting Deep_hour using string split method

df['Deep_hour']=df['Dep_Time'].str.split(':').str[0]
df['Deep_min']=df['Dep_Time'].str.split(':').str[1]
```

```python
# Dropping the Deep_hour column
df['Deep_hour']=df['Deep_hour'].astype(int)
df['Deep_min']=df['Deep_min'].astype(int)
```

```python
# converting the columns into int type
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(3)
```

```python
# Dropping the Route column
df.drop('Route',axis=1,inplace=True)
```

```python
# To find the number of unique stops in Total_Stops
df['Total_Stops'].unique()
```

```python
# Mapping all the unique stops with an integer
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head(3)
```

```python
# Splitting Deep_hour using string split method

df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
# Handling the '5m' exceptaion by droping it from the dataset(6474, this row no has no use in the dataset)
df[df['Duration_hour']=='5m']
```

```python
df.drop(6474, axis=0,inplace=True)
```

```python
df['Duration_hour']=df['Duration_hour'].astype(int)
```
